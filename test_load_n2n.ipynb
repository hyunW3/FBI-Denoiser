```python
def tensor_to_numpy(img):
    return img.detach().cpu().numpy()
```

```python
from matplotlib import pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr
t4_loader = loader_for_N2N(data_path="/mnt/ssd/hyun/fbi-net/FBI-Denoiser/data/train_Samsung_SNU_patches_230414.hdf5",
                        avg_num= 4, return_img_info=True)
for img_info, noisy1,noisy2, target_data_sample_tmp in t4_loader:
    noisy1 = tensor_to_numpy(noisy1)
    noisy2 = tensor_to_numpy(noisy2)
    target_data_sample_tmp = tensor_to_numpy(target_data_sample_tmp)
    print(img_info)
    print(noisy1.shape, noisy2.shape, target_data_sample_tmp.shape)
    plt.subplot(1,3,1)
    plt.title("PSNR : %.2f" % psnr(noisy1[0,:,:],target_data_sample_tmp[0,:,:],data_range=1.0))
    plt.imshow(noisy1[0,:,:],cmap="gray")
    plt.subplot(1,3,2)
    plt.title("PSNR : %.2f" % psnr(noisy2[0,:,:],target_data_sample_tmp[0,:,:],data_range=1.0))
    plt.imshow(noisy2[0,:,:],cmap="gray")
    plt.subplot(1,3,3)
    plt.imshow(target_data_sample_tmp[0,:,:],cmap="gray")
    
    break
```
