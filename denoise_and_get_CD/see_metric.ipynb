# PSNR / SSIM metric

```python
import numpy as np
metric = np.load('./PSNR_and_SSIM_metric.npy',allow_pickle=True).item()
metric
```

```python
metric['F#-F02_v2'].keys()
```

```python
refined_metric = {}
for method in metric.keys():
    refined_metric[method] = {}
    print(method)
    target = method.split("-")[1].split('_')[0]
    input = method.split("-")[0]
    print("====== ",method," ======")
    for key, values in metric[method].items():
        print(f"{key} - {np.mean(values):.2f} +- {np.std(values):.2f}")
        refined_metric[method][f"{key}_mean"] = np.mean(values)
        refined_metric[method][f"{key}_std"] = np.std(values)
    refined_metric[method]['target'] = target
    refined_metric[method]['input'] = input
```

```python
sort_metric = 'SSIM'
refined_metric = dict(sorted(refined_metric.items(), key = lambda x : x[1][f'{sort_metric}_mean'], reverse=True))
```

```python
for method in refined_metric.keys():
    print(f"{method}\t: {refined_metric[method][f'{sort_metric}_mean']:.2f} +- {refined_metric[method][f'{sort_metric}_std']:.2f}")
```

```python
import pandas as pd
df = pd.DataFrame(refined_metric).T
df
```

```python
df.to_csv(f"PSNR_and_SSIM_metric_in-terms-of-{sort_metric}.csv")
```
