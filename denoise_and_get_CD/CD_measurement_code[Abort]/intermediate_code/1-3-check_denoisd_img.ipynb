```python
import numpy as np
import os
from glob import glob
import matplotlib.pyplot as plt
import cv2
```

```python
data_path = glob("denoised_img_*")
```

```python
index = 1
for dataset_path in data_path:
    print(f"==== {dataset_path} ====")
    for set_num in sorted(os.listdir(dataset_path)):
        set_path = os.path.join(dataset_path,set_num)
        set_number = set_num[3:]
        for file_name in sorted(os.listdir(set_path)):
            print(set_num,file_name)
            f_num = file_name.split(".")[0]
            img_path = os.path.join(set_path,file_name)
            img = np.load(img_path)
            plt.title(f"{set_num} _ {f_num} {index}th image")
            img_uint8 = (img)#.astype('uint8')
            print(img_uint8.min(), img_uint8.max(),img_uint8.shape)
            plt.imshow(img_uint8[index][0])
            plt.pause(0.01)
            break
        break
```
