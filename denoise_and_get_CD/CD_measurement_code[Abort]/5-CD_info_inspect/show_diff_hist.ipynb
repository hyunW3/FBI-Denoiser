```python
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
path = "./CD_diff_info_F16_v2_150vsmedian_filter.txt"
with open(path, 'r') as f:
    CD_info = json.load(f)
```

```python
CD_info.keys()
```

```python
for idx,key in enumerate(CD_info.keys()):
# key = 'min_CD_diff'
    bins = np.linspace(min(CD_info[key]), max(CD_info[key]),100)
    plt.subplot(1,2,idx+1)
    plt.title(f"{key} (F16_v2 - median) ")
    plt.hist(CD_info[key], bins=bins)
    plt.axvline(-1,color='r')
    plt.axvline(1,color='r')
    print("")
    # plt.pause(0.01)
```
