```python
def boxlpot_with(refined_metric,metric_name,value_name,df):
    import matplotlib.pyplot as plt
    plt.figure(figsize=(20,10))
    plt.title(f"{metric_name} absolute difference (reference F32 median filterd)")
    plot = refined_metric.boxplot(column=list(refined_metric.columns))
    plt.ylim(0.0,3.0)
    plt.ylabel(f'{metric_name} absolute difference')

    plot.plot()
    plt.axhline(0.5,linestyle='--')
    plt.axhline(1.0,linestyle='--',color='r')

    plt.twinx()
    plt.plot([0] +list( df.keys()),[None] +list(df.loc[f'{value_name}_mean']))
    plt.ylabel(value_name)
    plt.legend([value_name])
    plt.show()
    plt.pause(0.01)
def make_boxplot(metric,metric_name = 'MAX CD'):
    psnr_df = pd.read_csv("../PSNR_and_SSIM_metric_in-terms-of-PSNR.csv")
    
    # add PSNR/SSIM value
    a = dict(metric)
    iter_key = list(a.keys())
    for key in iter_key:
        original_key =key
        while key[0] == '*':
            key = key[1:]
        if original_key != key :
            a.pop(original_key)
            a[key] = maxCD[original_key]
    for i in range(psnr_df.shape[0]):
        psnr_info = psnr_df.iloc[i]
        key = psnr_info[0]
        psnr_mean = psnr_info['PSNR_mean']
        ssim_mean = psnr_info['SSIM_mean']
        # print("=======",key, psnr_mean,ssim_mean)
        if key in a.keys():
            print("add psnr_mean in ",key,psnr_mean)
            a[key]['PSNR_mean'] = psnr_mean
            a[key]['SSIM_mean'] = ssim_mean
    # pd.DataFrame(a)
    # drop bad metric
    a.pop('F08-F32_v2')
    a.pop('F16-F32_v2')
    # dataframe with PSNR/SSIM & sort by PSNR
    refined_metric = dict(sorted(a.items(), key = lambda x : x[1][f'PSNR_mean'], reverse=True))
    refined_metric = pd.DataFrame(refined_metric)
    refined_metric = refined_metric.applymap(abs)
    
    # extract PSNR/SSIM dataframe
    psnr_df = refined_metric.loc['PSNR_mean']
    psnr_df = pd.DataFrame(psnr_df).T
    ssim_df = refined_metric.loc['SSIM_mean']
    ssim_df = pd.DataFrame(ssim_df).T
    
    # drop PSNR/SSIM to show boxplot
    refined_metric = refined_metric.drop(['PSNR_mean','SSIM_mean'])

    boxlpot_with(refined_metric,metric_name,'PSNR',psnr_df)
    boxlpot_with(refined_metric,metric_name,'SSIM',ssim_df)
    
```

```python
make_boxplot(maxCD,'MAX CD')
```
