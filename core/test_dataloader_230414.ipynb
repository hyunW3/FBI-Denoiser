```python
import numpy as np
from dataloader_230414 import RN2N_Dataset
```

```python
import torch
gen = RN2N_Dataset(x_avg_num=1,y_avg_num=32, return_img_info=True)
gen = torch.utils.data.DataLoader(gen, batch_size=1, shuffle=True, num_workers=1)
    
```

```python
from matplotlib import pyplot as plt
for idx, (img_info,noisy1,noisy2,clean) in enumerate(gen):
    print(img_info, noisy1.shape, noisy2.shape,clean.shape)
    plt.subplot(131)
    plt.imshow(noisy1[0,0,:,:].cpu().numpy(),cmap="gray")
    plt.subplot(132)
    plt.title(f'{img_info["img_name"][0]}_{img_info["pair_idx"].item()}th_(F{img_info["avg_num"][0].item()}/F{img_info["avg_num"][1].item()})')
    plt.imshow(noisy2[0,0,:,:].cpu().numpy(),cmap="gray")
    plt.subplot(133)
    plt.imshow(clean[0,0,:,:].cpu().numpy(),cmap="gray")
    plt.pause(0.01)
    if idx > 3:
        break
```
