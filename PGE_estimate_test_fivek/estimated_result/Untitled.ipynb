```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
parameters = [(0.01,0.0002),(0.01,0.02),(0.05,0.02)]
for test_alpha, test_beta in parameters:
    te_data_dir = '../test_fivek_rawRGB_alpha_'+str(test_alpha)+'_beta_'+str(test_beta)+'.hdf5'
    with h5py.File(te_data_dir,'r') as f:
        noisy_images = np.array(f['noisy_images'])
        clean_images = np.array(f['clean_images'])
    for index in [193]:
        plt.subplot(121)
        plt.title(f"noisy_images\n{test_alpha}, {test_beta}\n {index}th  image")
        plt.imshow(noisy_images[index])
        plt.subplot(122)
        plt.title(f"clean_images\n{test_alpha}, {test_beta}\n {index}th  image")
        plt.imshow(clean_images[index])
        plt.pause(0.01)
    print(noisy_images.shape,clean_images.shape)
```
